# Le funzioni $F: \mathbf{R}^N \rightarrow \mathbf{R}^M$

## Esempio: coordinate sferiche

Sia $F: \mathbf{R}^2 \rightarrow \mathbf{R}^3$ definita nel seguente modo:

$$
F(\theta, \phi) =
\begin{cases}
y_1 &= \rho \sin(\theta) \cos (\phi)\\
y_2 &= \rho \sin(\theta) \sin(\phi)\\
y_3 &= \rho \cos(\theta)
\end{cases}
$$

con $\rho$ fissato (ad es. $\rho = 1$), $0 \leq \theta \leq \pi$ e $0 \leq \phi \leq 2 \pi$

```python
import numpy as np
import matplotlib.pyplot as plt

RHO = 1
```

$f_1 (\theta, \phi) = \rho \sin(\theta) \cos(\phi)$

```python
def F_1(x):
    return RHO * np.sin( x[0] ) * np.cos( x[1] )
```

$f_2(\theta, \phi) = \rho \sin(\theta) \sin(\phi)$

```python
def F_2(x):
    return RHO * np.sin( x[0] ) * np.sin( x[1] )
```

$f_3(\theta, \phi) = \rho \cos(\theta)$

```python
def F_3(x):
    return RHO * np.cos( x[0] )
```

```python
def F(x):
    return [F_1(x), F_2(x), F_3(x)]
```

I nostri vettori di input saranno $\underline{x} = (\theta, \phi)$ in $\mathbf{R}^2$.

Creiamo una meshgrid:

```python
theta = np.linspace(0, np.pi, 20) # latitude
phi = np.linspace(0, 2 * np.pi, 20) # longitude

x = np.meshgrid(theta, phi)
```

```python
plt.scatter(*x)
plt.show()
```

Per ogni punto della meshgrid, calcoliamo la nostra funzione:

```python
ax = plt.axes(projection='3d')

ax.plot_wireframe( *F(x), color='r' ) # wireframe( F_1, F_2, F_3 ) *[F_1, F_2, F_3]

plt.show()
```

```python
ax = plt.axes(projection='3d')

ax.plot_surface( *F(x), color='b' )

plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.plot_wireframe( *F(x), color='b' )

ax.contourf( *F(x), zdir='z', offset=-1, cmap='coolwarm')
ax.contourf( *F(x), zdir='x', offset=-1, cmap='coolwarm')
ax.contourf( *F(x), zdir='y', offset=1, cmap='coolwarm')

plt.show()
```

## Esempio: $f$ continua separatamente in $x$ e $y$

Sia $f: \mathbf{R}^2 \rightarrow \mathbf{R}$, definita nel seguente modo:

$$
f(x, y) =
\begin{cases}
\frac{xy}{x^2 + y^2} & \text{se } (x, y) \not = (0, 0)\\
0 & \text{se } (x, y) = (0, 0)
\end{cases}
$$

```python
def f(x, y):
    return ( x * y ) / ( x**2 + y**2 )
```

Creiamo la meshgrid:

```python
N = 51

lin_x = np.linspace(-1, 1, N)
lin_y = np.linspace(-1, 1, N)

x, y = np.meshgrid(lin_x, lin_y)
```

```python
plt.scatter(x, y, s=1)
plt.scatter(0, 0, s=30, facecolors='none', edgecolors='r')

# plt.grid(True)

plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.scatter(x, y, f(x, y))
```

```python
from matplotlib import cm

ax2 = plt.axes(projection='3d')
ax2.plot_surface(x, y, f(x, y), cmap=cm.viridis)
```

```python
ax3 = plt.axes(projection='3d')
# ax3.plot_surface(x, y, f(x, y), cmap=cm.viridis)

ax3.contourf( x, y, f(x, y), zdir='z', offset=-0.5, cmap='coolwarm')
ax3.contourf( x, y, f(x, y), zdir='x', offset=-1, cmap='coolwarm')
ax3.contourf( x, y, f(x, y), zdir='y', offset=1, cmap='coolwarm')
```

Sia $f: \mathbf{R}^2 \rightarrow \mathbf{R}$, definita nel seguente modo:

$$
f(x, y) =
\begin{cases}
{\left( \frac{x^2 y}{x^4 + y^2} \right)}^2 & \text{se } (x, y) \not = (0, 0)\\
0 & \text{se } (x, y) = (0, 0)
\end{cases}
$$

```python
def f2(x, y):
    return ( (x**2 * y) / (x**4 + y**2) )**2
```

```python
ax2 = plt.axes(projection='3d')
ax2.plot_surface(x, y, f2(x, y), cmap=cm.viridis)
```
